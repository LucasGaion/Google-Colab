```python
import numpy as np
import gym
import random
from collections import deque
import tensorflow as tf
from tensorflow import keras

# Parâmetros
EPISODES = 1000
REPLAY_MEMORY_SIZE = 2000
BATCH_SIZE = 32
DISCOUNT = 0.95
EPSILON = 1.0
EPSILON_MIN = 0.01
EPSILON_DECAY = 0.995

# Criação do ambiente
env = gym.make('CartPole-v1')

# Construção da rede neural
model = keras.Sequential([
    keras.layers.Dense(24, input_dim=4, activation='relu'),
    keras.layers.Dense(24, activation='relu'),
    keras.layers.Dense(2, activation='linear')
])
model.compile(optimizer='adam', loss='mse')

# Replay memory
replay_memory = deque(maxlen=REPLAY_MEMORY_SIZE)

def get_action(state):
    if np.random.rand() <= EPSILON:
        return random.randrange(2)  # Ação aleatória
    q_values = model.predict(state)
    return np.argmax(q_values[0])  # Ação com o maior Q-value

# Loop de treinamento
for episode in range(EPISODES):
    state = env.reset()
    state = np.reshape(state, [1, 4])
    total_reward = 0

    for time in range(500):
        action = get_action(state)
        next_state, reward, done, _ = env.step(action)

        # Adiciona a experiência à memória
        next_state = np.reshape(next_state, [1, 4])
        replay_memory.append((state, action, reward, next_state, done))

        state = next_state
        total_reward += reward

        # Treinamento da rede neural
        if len(replay_memory) > BATCH_SIZE:
            mini_batch = random.sample(replay_memory, BATCH_SIZE)
            for s, a, r, ns, d in mini_batch:
                target = r
                if not d:
                    target += DISCOUNT * np.amax(model.predict(ns)[0])
                target_f = model.predict(s)
                target_f[0][a] = target
                model.fit(s, target_f, epochs=1, verbose=0)

        if done:
            break

    # Redução do epsilon
    if EPSILON > EPSILON_MIN:
        EPSILON *= EPSILON_DECAY

    print(f'Episódio: {episode+1}, Total de Recompensas: {total_reward}')

env.close()
```
